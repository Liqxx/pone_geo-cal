```python
import os
import copy
import numpy as np
import scipy.stats as stat
import matplotlib.pyplot as plt
```

```python
def corner(z, title='Test'):
    plot_xl = ['abs', 'sca', 'domeff', 'p0', 'p1']

    fig = plt.figure(figsize=(16,8))

    axs  = []
    cntr = 1
    for i in range(d):
        for j in range(d):

            # diagonals
            if i == j:
                ax = plt.subplot(d, d, cntr)
                axs.append(ax)
                ax.hist(z[:,i], histtype='step')

                # add labels
                if i == 0:
                    ax.set_ylabel('Counts')
                if i == d - 1:
                    ax.set_xlabel(plot_xl[i]) # or dim2, dim1 = dim2 for diagonals

            # off diagonals
            if i > j:
                ax = plt.subplot(d, d, cntr, sharex=axs[j])
                ax.plot(z[:,j], z[:,i], ls='', marker='o', ms=3)

                # add labels
                # to y-axis on leftmost column
                if j == 0:
                    ax.set_ylabel(plot_xl[i])
                # to x-axis on lowest row
                if i == d - 1:
                    ax.set_xlabel(plot_xl[j])

            cntr += 1

    fig.suptitle(title)
    plt.show()
```

# Comparison: Numpy Random vs. Rnew

```python
# number of dimensions
d = 5
```

```python
# number of points
n = 500
```

```python
# quasi-random sampling
import quasi_random_rng
reload(quasi_random_rng)

z = quasi_random_rng.rnew(d, n)
```

```python
# numpy random uniform
znumpy = np.random.rand(n, d)
```

```python
corner(z, title='R2')
```

```python
corner(znumpy, title='Numpy Random')
```

# Using priors

```python
# use normal prior for abs, sca, domeff
# centered at 1, sigma 0.3
# truncate on +- 2 sigma
loc    = 1
scale  = 0.3
clip_a = loc - 2 * scale
clip_b = loc + 2 * scale

# get truncated interval
a, b   = (clip_a - loc) / scale, (clip_b - loc) / scale

# get prior distribution from uniform
# by using inverse CDF
# explanation: CDF transforms distribution to [0,1]
# inverse transforms back
zprior = stat.truncnorm(a, b, loc=loc, scale=scale).isf(z)
```

```python
znew = copy.deepcopy(z)

znew[:,0] = zprior[:,0]
znew[:,1] = zprior[:,1]
znew[:,2] = zprior[:,2]

# replace abs, sca, and domeff priors 
znew[:,0] = zprior[:,0]
znew[:,1] = zprior[:,1]
znew[:,2] = zprior[:,2]

# set p0 range to [-1, 1] uniform
a = - 1.0
b = 1.0
znew[:,3] = z[:,3] * (b - a) + a

# set p1 range to [-0.2, 0.2] uniform
a = -0.2
b = 0.2
znew[:,4] = z[:,4] * (b - a) + a
```

```python
corner(znew, title='Truncated normal prior R2')
```
