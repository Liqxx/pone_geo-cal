```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="0"

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

%load_ext autoreload
%autoreload 2
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = [12., 8.]
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14 
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['legend.fontsize'] = 14
```

```python
A = np.load('x_10runs.npy').squeeze()
B = np.load('theta_10runs.npy').squeeze()
```

```python
data = np.hstack([A, B])
df = pd.DataFrame(data)
df.columns = ['x%s'%i for i in range(20)] + ['pocam', 'string', 'om', 'dist', 'zenith','scat', 'dom_eff']
```

```python
# the data
df
```

```python
# inputs and outputs for the NN
t_columns = ['dom_eff', 'scat', 'pocam', 'dist', 'zenith']
x_columns = df.columns[:20]
```

```python
x = df[x_columns].values
t = df[t_columns].values
x_train, x_test, t_train, t_test = train_test_split(x, t, test_size=0.2, random_state=42)
```

```python
activation='relu'

inputs = tf.keras.Input(shape=(t_train.shape[1],))
h = tf.keras.layers.Dense(64, activation=activation)(inputs)
#h = tf.keras.layers.Dense(64, activation=activation)(h)
h = tf.keras.layers.Dense(256, activation=activation)(h)
h = tf.keras.layers.Dense(256, activation=activation)(h)
#h = tf.keras.layers.Dense(512, activation=activation)(h)
outputs = tf.keras.layers.Dense(x_train.shape[1], activation='softplus')(h)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(0.001)
model.compile(loss='poisson', optimizer=optimizer)
```

```python
hist = model.fit(t_train, x_train, epochs=30, verbose=1, validation_data=(t_test, x_test), batch_size=4096)
```

```python
# make a few random plots for a random datapoints

fig, ax = plt.subplots(4,4, figsize=(20,20))

ax = ax.flatten()

for j in range(16):

    axis = ax[j]
    rand_idx = np.random.randint(0, t_test.shape[0])
    inp = t_test[rand_idx]
    outp = x_test[rand_idx]

    pred = model.predict(inp[np.newaxis,:])[0]

    axis.plot(outp, label='Data')
    axis.plot(pred, label='NN expectation')
ax[-1].legend()
```

```python
# interpolation behaviour for dom efficiency

N = 51

rand_idx = np.random.randint(0, t_test.shape[0])
inp = t_test[rand_idx]

inp = np.repeat(inp[np.newaxis,:], N, axis=0)

col = t_columns.index('dom_eff')

inp[:, col] = np.linspace(0.5, 1.5, N)

   
ps = model.predict(inp)

colors = plt.cm.jet(np.linspace(0,1,N))
for i, (c, p) in enumerate(zip(colors, ps)):
    label = 'dom eff = %.2f'%inp[i,col] if i%10==0 else None
    alpha = 1 if i%10==0 else 0.25
    plt.plot(p, c=c, label=label, alpha=alpha)
plt.legend()
```

```python
# make a few plots
dom_effs = df.dom_eff.unique()
scat = 1.0
pocam = 7.0
string = 79
om = 27

colors = plt.cm.jet(np.linspace(0,1,len(dom_effs)))
fig, ax = plt.subplots(4,4, figsize=(20,20))
ax = ax.flatten()

for j in range(16):
    
    om += 1
    axis = ax[j]
    for i, dom_eff in enumerate(dom_effs):

        sf = df.loc[(df.dom_eff == dom_eff) & (df.scat == scat) & (df.pocam == pocam) & (df.string == string) & (df.om == om)]
        axis.plot(model.predict(sf[t_columns].values[[0]])[0], c=colors[i], label='Dom eff. = %.2f'%dom_eff)
        axis.plot(sf[x_columns].values.T, c=colors[i], ls=':', alpha=0.3)
        axis.plot(np.average(sf[x_columns].values, axis=0), c=colors[i], ls='--')
        axis.set_title('OM %i'%om)

ax[-1].legend()
plt.savefig('test.png')
```
